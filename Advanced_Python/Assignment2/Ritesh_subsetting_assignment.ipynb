#  importing the libraries
# reading the netflix file into dataframe











```python
import numpy as np
import pandas as pd

df=pd.read_excel('netflix.xlsx')
```

```python
df.info()
```

# 1.    Select TV Show released in year 2019

```python
df[(df.release_year == 2019) & (df.type=='TV Show')]
```

# 2.    Select Movies between year 2018 to 2020

```python
df[((df.release_year >= 2018) & (df.release_year <= 2020)) & (df.type=='Movie')]
```

# 3.    Select shows aired only in United states

```python
df[df.country=='United States']
```

# 4.    Select cast of the TV shows  telecasted in year 2019 and had rating R

```python
df[(df.release_year==2019) & (df.rating=='R')]['cast']
```

# 5.    Drop column show_id

```python
df.drop(['show_id'], axis = 1, inplace = True)
```

```python
df.info()
```

# 6.    Add column duration_min such that it’s type is numeric [use column duration]

```python
df['duration_min'] = df.duration.apply(lambda x : x.split()[0])
```
